```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
hr = pd.read_excel("HR Cleaned Dataset1.xlsx")
```

### Employees Distribution by Gender

```python
plt.figure(figsize=(4,4))
sns.countplot(hr, x="GenderID", palette="muted")
plt.title("Employees Analysis by Gender")
plt.show()
```

From the above graph we can see that most of the employees are Females.

```python
position_sal = hr.groupby("Position")["Salary"].sum().reset_index()
```

```python
position_sal
```

### Salary Distribution By Department using BarPlot

```python
plt.figure(figsize=(12,8))
hx = sns.barplot(position_sal, x="Salary", y="Position", palette="flare", hue="Salary")
plt.show()
```

From above graph we can see that Product Technician I and Product Technician II have the highest salary.

### Performance Score of Employees by Race

### Employee Count based on their Marital Status

```python
plt.figure(figsize=(6,4))
hx = sns.countplot(hr, x="Marital Status", hue="GenderID", palette="mako")
for bars in hx.containers:
    hx.bar_label(bars)
plt.show()
```

From above graph we can see that most of the employees are Single.

### Employee Satisfaction Score comparing to their Termination Reason

```python
plt.figure(figsize=(10,4))
sns.swarmplot(hr, x="EmpSatisfaction", y="Termination Reason", hue="EmpSatisfaction", palette="viridis")
plt.title="Employee Termination Analysis"
plt.show()
```

From above graph we can see that termination reason contributes major role in employee satisfaction score.

### Employee Count by Recruitment Source

```python
plt.figure(figsize=(16,8))
hx = sns.countplot(hr, x="RecruitmentSource", palette="mako")
for bars in hx.containers:
    hx.bar_label(bars)
plt.show()
```

From above graph we can see higher number of employees are selected from "Indeed" recruitment source.

### Salary Distributiob of employees by 

```python
Dept_sal = hr.groupby("Department").agg(Agg_salary=('Salary', 'sum'),
    Employee_Count=('Employee_Name', 'count')).reset_index()
Dept_sal
```

```python
plt.figure(figsize=(12,6))
hx = sns.barplot(data=Dept_sal, x="Employee_Count", y="Department", fill=True, label="Employee_Count", palette="flare")
for bars in hx.containers:
    hx.bar_label(bars)
plt.xlabel("Aggregate Salary")
plt.ylabel("Department")
plt.show()
```

From above graph we can see that Production Department has the highest Salary.

### Total Employee count By Department using Pie Chart

```python
plt.figure(figsize=(8, 8))
plt.pie(Dept_sal['Employee_Count'], labels=Dept_sal['Department'], labeldistance=1.05, autopct="%.2f%%",
       explode=[0, 0, 0, 0, 0, 0], startangle=40, 
        colors=["#cc0000","#ff0000","#ff3333","#ff6666","#ff9999","#ffcccc"])
plt.show()
```

From above Pie chart we can see that Production Department has the highest number of employees.

### Late employees by Department

### Absences and Age analysis based on Department by KDE plot

```python
plt.figure(figsize=(16,8))
plt.subplot(1, 2, 1)
sns.kdeplot(hr, x="Absences", fill=True, hue="Department", palette="rocket")
plt.subplot(1, 2, 2)
sns.kdeplot(hr, x="Age", fill=True, hue="Department", palette="rocket")
plt.show()
```

From above graphs we can see that Production department has the higher number of absences from age group 40-60.

### Employee Satisfaction Distribution by using Histplot

```python
plt.figure(figsize=(8,4))
hx = sns.histplot(hr, x="EmpSatisfaction", bins=20, kde=True, cumulative=True, color="purple")
for bars in hx.containers:
    hx.bar_label(bars)
plt.show()
```
